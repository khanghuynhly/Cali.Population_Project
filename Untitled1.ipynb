# Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_cities = pd.read_csv('california_cities.csv')
df_cities.head()
```

```python
x = df_cities['longd']
y = df_cities['latd']
pop = df_cities['population_total']
are = df_cities['area_total_sq_mi']
plt.figure(figsize=(8,6))
plt.scatter(x, y, 
           c=np.log10(pop), cmap='viridis',
           s=are, linewidths=0, alpha=0.5);
plt.ylabel("latitude")
plt.xlabel("longitude")
plt.colorbar(label='log$_{10}$(population)')
plt.axis("tight")
plt.style.use('_classic_test_patch')
plt.title("California Cities: Population and Area Distribution")
```
